```python
# load without headers to inspect the top rows
df_raw = pd.read_csv("Marital.csv", header=None)

#show the first 10 rows to understand the structure
df_raw.head(10)
```

```python
row_0 = df_raw.iloc[0] # Original names (may be long/incomplete)
row_2 = df_raw.iloc[2] # Category headers ("national level", etc.)

# Fill missing row_0 values with placeholders to avoid NaNs
row_0_filled = row_0.fillna("Unknown")

# Same for row_2 (in case some are empty)
row_2_filled = row_2.fillna("")

#Combine the two with hyphen (or underscore if preferred)
combined_headers = row_0_filled + " - " + row_2_filled
```

```python
df = df_raw.iloc[3:].copy()
df.columns = combined_headers
df.reset_index(drop=True, inplace=True)
```
